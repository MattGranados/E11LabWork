```python
import math
import numpy as np
import pandas as pd
# This line allows plots to show in the Jupyter notebook
%matplotlib inline
import matplotlib.pyplot as plt
import io
import requests
from scipy.stats import norm
from IPython import display
```

```python
OurData = pd.read_csv('GPS_GUI_Data_2021-11-17_AR.csv')
Data1 = pd.read_csv('Map-data-0.csv')
Data2 = pd.read_csv('Map-data-1.csv')
```

```python
# Part A - position not plotted since data does not include position daata
plt.plot(Data1['Radiation (cps)'])
plt.plot(Data2['Radiation (cps)'])
plt.ylabel('Number of Counts') 
plt.xlabel('Time')                    
plt.title("Counts vs Time for provided Data")                                 
plt.show()
```

The two data sets look relatively similar. There are differences between the two graphs, but for the most part, the deviations from each other appear, visually speaking, to not be too far off from each other

```python
## Part B
plt.ion()
fig1 = plt.figure(1, figsize=(10,5))
ax1 = fig1.add_subplot()
fig1.canvas.draw()
ax1.clear()
plt.hist(Data1['Radiation (cps)'],bins = 10)
ax1.set_ylabel("Frequency")
ax1.set_xlabel("Counts")
fig1.canvas.draw()
display.display(plt.gcf())
display.clear_output(wait=True)
mu1 = np.mean(Data1['Radiation (cps)'])
sig1 = np.std(Data1['Radiation (cps)'])
theory1 = math.sqrt(mu1)
```

```python
plt.ion()
fig1 = plt.figure(1, figsize=(10,5))
ax1 = fig1.add_subplot()
fig1.canvas.draw()
ax1.clear()
plt.hist(Data2['Radiation (cps)'],bins = 10)
ax1.set_ylabel("Frequency")
ax1.set_xlabel("Counts")
fig1.canvas.draw()
display.display(plt.gcf())
display.clear_output(wait=True)
mu2 = np.mean(Data2['Radiation (cps)'])
sig2 = np.std(Data2['Radiation (cps)'])
theory2 = math.sqrt(mu2)
```

```python
print('The mean and standardard deviation of the first data set are {:.5} and {:.5} respectively'.format(mu1,sig1))
print('The mean and standardard deviation of the second data set are {:.5} and {:.5} respectively'.format(mu2,sig2))
```

Statistically, the data sets agree with each other. The standard deviations are incredibly close to each other and the means are only about 2 counts off from each other. This 2 count offset is within the 4.6/4.7 standard deviations of the data sets, further indicating that the data sets agree with each other. 

```python
## Part C
print('The counting statisitics variance is {:.3} for the first data set'.format(theory1))
print('The counting statisitics variance is {:.3} for the second data set'.format(theory2))
```

According to counting statisitcs, a variation of 5.62/5.8 counts was expected; this is about a count higher than the actual standard deviations of 4.62/4.8

```python
## Part D
Uncertain = math.sqrt(mu1)/(math.sqrt(2*math.sqrt(155)))
print('The expected uncertainty on the first data set is {:.3}'.format(Uncertain))
Uncertain2 = math.sqrt(mu2)/(math.sqrt(2*math.sqrt(154)))
print('The expected uncertainty on the second data set is {:.3}'.format(Uncertain2))
```

# Part E
Comparing the calculated variances with uncertainties and the calculated standard deviations, the first data set's variation is what was expected from simple counting statisitics. For the second data set. the variation is approximatelt what was expected; although the standard deviation is just every so slightly outside the variance + uncertainty. This tells us that the sources of radiation contributing to the data generally follow a normal distribution and that the source did not appear to change as more measurements were taken.
